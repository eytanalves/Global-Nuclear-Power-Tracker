```python
import folium
from folium.plugins import MarkerCluster
from branca.element import Template, MacroElement

template = """
{% macro html(this, kwargs) %}
<div style="
    position: fixed; 
    bottom: 50px;
    left: 50px;
    width: 150px;
    height: 250px; 
    background-color: white;
    border:2px solid grey;
    z-index:1000;
    font-size:14px;
    ">
    <p><a style="color:green;font-size:150%;margin-left:20px;">•</a>&emsp;Operating</p>
    <p><a style="color:red;font-size:150%;margin-left:20px;">•</a>&emsp;Retired</p>
    <p><a style="color:orange;font-size:150%;margin-left:20px;">•</a>&emsp;Construction</p>
    <p><a style="color:pink;font-size:150%;margin-left:20px;">•</a>&emsp;Pre-Construction</p>
    <p><a style="color:blue;font-size:150%;margin-left:20px;">•</a>&emsp;Announced</p>
    <p><a style="color:gray;font-size:150%;margin-left:20px;">•</a>&emsp;Others</p>
</div>
{% endmacro %}
"""


macro = MacroElement()
macro._template = Template(template)

map.get_root().add_child(macro)



def color(status):
    if status == 'operating':
        return 'green'
    elif status == 'retired':
        return 'red'
    elif status == 'construction':
        return 'orange '
    elif status == 'pre-construction':
        return 'pink'
    elif status == 'announced':
        return 'blue'
    else:
        return 'gray'



map = folium.Map(location=[0, 0], zoom_start=2)

map.get_root().add_child(macro)

marker_cluster = MarkerCluster().add_to(map)


for index, row in nuclear_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=folium.Icon(color=color(row['Status'])),
        popup=f"<b>Model:</b> {row['Model']}<br><b>Status:</b> {row['Status']}<br><b>Country:</b> {row['Country']}\
        <b>Start Year:<br>{row['Start Year']}\n"
        +f'<b>Retired Year:</b> {row["Retired Year"]}' if row['Status'] == 'retired' else f"<b>Model:</b> {row['Model']}<br><b>Status:</b> {row['Status']}<br><b>Country:</b> {row['Country']}\
        <b>Start Year:<br>{row['Start Year']}\n"  
    ).add_to(marker_cluster)


# folium.LayerControl().add_to(map)


map
```
